#### Initial data clean-up and set up. Read csv files into pandas dataframe to get required content and pivot as needed to create SQLite DB. 


```python
# Dependencies to create SQLlite database
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
from sqlalchemy import Column, Integer, String, Float, Text, ForeignKey
```

```python
# Create an engine to a SQLite database 
engine = create_engine("sqlite:///belly_button_biodiversity.sqlite")
```

```python
# Create a connection to the engine called `conn`
conn = engine.connect()
```

```python
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base

Base.classes.keys()
```

```python

# Assign the class to a variable 

otu = Base.classes.otu
samples = Base.classes.samples
samples_metadata= Base.classes.samples_metadata
```

```python
# Create a session

session = Session(engine)
```

```python
for row in session.query(samples_metadata).limit(5).all():
    print(row)
```

```python
otu_results = session.query(otu.lowest_taxonomic_unit_found).all()
otu_results
```

```python
metadata = session.query(samples_metadata.SAMPLEID, samples_metadata.ETHNICITY, samples_metadata.GENDER, samples_metadata.AGE, samples_metadata.LOCATION,samples_metadata.BBTYPE).all()
metadata
```
